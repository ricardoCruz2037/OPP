## Diccionarios
A diferencia de las listas se declaran con las llaves:  dic = {}, los diccionarios llevan claves, palabra clave separada por dos puntos: dic = {'Nombre': 'Richi'}

```python
dic = {'Nombre': 'Richi', 'Edad': 19,'Altura': 1.65}
print(dic)
```

Hay momentos donde en un for no quiere accesar a con un indice para recorrer uno por uno

```python
list = [2, 3, -5, 4]
print(list)

for valor in list:
    print(valor)
```

Y como recorro un diccionario?

```python
dic = {'Nombre': 'Richi', 'Edad': 19,'Altura': 1.65}
print('Dic', dic,'\n')

# Al recorrer diccionario
for valor in dic:
    print(valor)
    
print('\n')
# Al recorrer keys
for valor in dic.keys():
    print(valor)
    
print('\n')

# Al recorrer values
for valor in dic.values():
    print(valor)
    
# Imprimir key en especifico
print('\n',dic['Nombre'])

# Imprimir propiedades
print(dir(dic))

# Imprimir Help
print('\n',help(dic))

# Regresa una t√∫pla de objetos
print('\n',dic.items(),'\n')

# Items, ambos
for valor in dic.items():
    print(valor)

print('\n')
    
# keys y valores
for key, valor in dic.items():
    print(key, valor)

dic['Nombre'] = 'Dan'
print(dic)
```

En una lista le pasas el numero de indice, en un diccionario le pasas la key!
